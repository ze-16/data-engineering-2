```python

image_dir = "preprocessed_images/"



feature_list = []


for filename in os.listdir(image_dir):
    if filename.endswith((".jpg", ".png", ".jpeg")):
      image_path = os.path.join(image_dir, filename)
      image = cv2.imread(image_path)

      mean_intensity = np.mean(image)
      norm_intensity =  np.linalg.norm(image)

      gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
      edges = cv2.Canny (gray_image,100,200 )
      contours,_ = cv2.findContours (edges, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

      shape_feature = []
      for contour in contours:
        area = cv2.contourArea (contour)
        perimeter = cv2.arcLength(contour,True)
        moments = cv2.moments(contour)
        if moments ["m00"] != 0:
          cx = int(moments["m10"]/moments["m00"])
          cy = int(moments["m01"]/moments["m00"])

        else:
          cx, cy = 0,0

        x,y,w,h = cv2.boundingRect(contour)

        shape_feature.append({
                "area": area,
                "perimeter": perimeter,
                "centroid": (cx, cy),
                "bounding_box": (x, y, w, h)
            })



      feature_extracted = {
          "mean_intensity" : mean_intensity,
          "norm_intensity" : norm_intensity,
          "shape_feeatures" :shape_feature
      }
      feature_list.append(feature_extracted)


with open ("image_features.json",'w') as json_file:
  json.dump(feature_list, json_file, indent=4)
```

```python
output_directory = "/content/processed_text_files"
os.makedirs(output_directory, exist_ok=True)
text_directory = "/content"

for filename in os.listdir(text_directory):
    if filename.endswith((".txt")):
      file_path = os.path.join(text_directory, filename)
      with open(filename, "r") as file:
        text = file.read()
        text = text.lower()
        token_text = word_tokenize(text)
        tokens_without_punctuations = [word for word in token_text if word not in string.punctuation]
        stop_words = set(stopwords.words("english"))
        filtered_tokens = [word for word in tokens_without_punctuations if word not in stop_words]
        print ("Preprocessed tokens:", filtered_tokens )


      output_path = os.path.join(output_directory, filename)
      with open(output_path, "w", encoding="utf-8") as output_file:
          output_file.write(" ".join(filtered_tokens))
          print(f"Preprocessed and saved: {output_path}")


```

```python

text_directory = "/content/processed_text_files"
documents = []


for filename in os.listdir(text_directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(text_directory, filename)
        with open(file_path, "r", encoding="utf-8") as file:
            documents.append({"filename": filename, "content": file.read()})


vectorizer_bow = CountVectorizer()
vectorizer_tfidf = TfidfVectorizer()


output_data = []


for i, document in enumerate(documents):
    bow_matrix = vectorizer_bow.fit_transform([document["content"]])
    tfidf_matrix = vectorizer_tfidf.fit_transform([document["content"]])


    doc_data = {
        "filename": document["filename"],
        "BoW": {
            "matrix": bow_matrix.toarray().tolist(),
            "vocabulary": vectorizer_bow.get_feature_names_out().tolist()
        },
        "TF-IDF": {
            "matrix": tfidf_matrix.toarray().tolist(),
            "vocabulary": vectorizer_tfidf.get_feature_names_out().tolist()
        }
    }
    output_data.append(doc_data)


output_file = "/content/vectorized_documents.json"
with open(output_file, "w", encoding="utf-8") as json_file:
    json.dump(output_data, json_file, ensure_ascii=False, indent=4)
```
